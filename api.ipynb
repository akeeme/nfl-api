```python
import requests
import pandas as pd
import json
```

```python
response = requests.get("https://www.espn.com/nfl/stats/player")
needed = response.text.split("playerStats")[1].split("views")[0]
```

```python
# clean needed, then turn into json
needed = needed[2:]
needed = needed[0:-2]
needed = needed.strip()
needed = json.loads(needed)
```
